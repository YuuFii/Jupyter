## 1. Persamaan Non-Linear

### a. Menggambar Grafik (Metode Grafis)

Gambar grafik dari fungsi f(x) = x<sup>3</sup>-30x<sup>2</sup>+2552 dengan interval [0,20]

```python
import numpy as np
import matplotlib.pyplot as plt

def plot():
    x = np.arange(0,20,1)
    fx = x**3-30*x**2+2552
    plt.plot(x,fx,'-')
    plt.grid(True)
    plt.xlabel('X')
    plt.ylabel('Fungsi f(x) = x^3+30x^2+2552')
    plt.show()

plot()
```

### b. Metode Biseksi

Menyelesaikan persamaan f(x) = x<sup>3</sup>-2x<sup>2</sup>+6x-4 menggunakan metode biseksi

```python
def biseksi():
    fx = lambda x: x**3 - 2*(x**2) + 6*x - 4
    a = float(input('Masukkan a: '))
    b = float(input('Masukkan b: '))
    
    print('a\tb\tc\tfx(a)\tfx(b)\tfx(c)')
    
    if fx(a) * fx(b) < 0:
        while abs(fx(a) * fx(b)) > 0.000001:
            c = (a + b) / 2
            print('{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}'.format(a, b, c, fx(a), fx(b), fx(c)))
            
            if fx(a) * fx(c) > 0:
                a = c
            else:
                b = c
        print('\nAkar persamaan adalah', round(c,3))
    else:
        print('Penentuan a dan b salah')

biseksi()
```

#### Contoh Soal

Terdapat fungsi f(x) = sin(10x) + cos(3x), dengan interval [3,6], nilai increment 0.01.

- Buat program untuk menggambar fungsi f(x)
- Cari nilai akarnya menggunakan program


```python
def plotContoh() :
    x = np.arange(3,6,0.01)
    fx = np.sin(10*x) + np.cos(3*x)
    plt.plot(x, fx, '-')
    plt.grid(True)
    plt.xlabel('X')
    plt.ylabel('Fungsi f(x) = sin(10x) + cos(3x)')
    plt.show()
    
def biseksiContoh() :
    fx = lambda x: np.sin(10*x) + np.cos(3*x)
    a = 3.5
    b = 4
    
    print('a\tb\tc\tfx(a)\tfx(b)\tfx(c)')
    
    if fx(a) * fx(b) < 0:
        while abs(fx(a) * fx(b)) > 0.000001:
            c = (a + b) / 2
            print('{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}'.format(a, b, c, fx(a), fx(b), fx(c)))
            
            if fx(a) * fx(c) > 0:
                a = c
            else:
                b = c
        print('\nAkar persamaan adalah', round(c,3))
    else:
        print('Penentuan a dan b salah')
    
plotContoh()
biseksiContoh()
```

- Ubah program agar evaluasi thresholdnya menggunakan rumus error relatif (Er) < 0.005

```python
import numpy as np

def biseksiContohUbah() :
    fx = lambda x: np.sin(10*x) + np.cos(3*x)
    a = float(input('Masukkan a :'))
    b = float(input('Masukkan b :'))
    Er = 10000.0
    cp = a
    
    if (fx(a) * fx(b)) < 0.0:
        n = 0
        print('Iterasi\t\ta\t\tb\t\tc\t\tfx(a)\t\tfx(b)\t\tfx(c)\t\tEr')
        
        while abs(Er) > 0.5:
            c = (a + b) / 2
            cc = c
            Er = ((cc-cp)/cc) * 100
            
            print('{:d}\t\t{:.6f}\t{:.6f}\t{:.6f}\t{:.6f}\t{:.6f}\t{:.6f}\t{:.6f}'.format(n, a, b, c, fx(a), fx(b), fx(c), abs(Er)))
            
            cp = cc
            n += 1
            
            if fx(a) * fx(c) > 0.0:
                a = c
            else:
                b = c
        print('\nAkar persamaan adalah', round(c,6))
    else:
        print('Penentuan a dan b salah')
    
biseksiContohUbah()
```

### c. Metode Regula Falsi

```python
import numpy as np

def regula_falsi() :
    fx = lambda x: x**3 - 2*(x**2) + 6*x - 4
    
    a = float(input('Masukkan a :'))
    b = float(input('Masukkan b :'))
    Er = 10000.0
    cp = a
    
    if (fx(a) * fx(b)) < 0.0:
        n = 0
        print('\nIterasi\t\ta\t\tb\t\tc\t\tfx(a)\t\tfx(b)\t\tfx(c)\t\tEr')
        while abs(Er) > 0.5:
            c = b - (fx(b) * (a - b))/(fx(a) - fx(b))
            cc = c
            Er = ((cc-cp)/cc) * 100
            
            print('{:d}\t\t{:.6f}\t{:.6f}\t{:.6f}\t{:.6f}\t{:.6f}\t{:.6f}\t{:.6f}'.format(n, a, b, c, fx(a), fx(b), fx(c), abs(Er)))
            
            cp = cc
            n += 1
            
            if fx(a) * fx(c) > 0.0:
                a = c
            else:
                b = c
        print('\nAkar persamaan adalah', round(c,6))
    else:
        print('Penentuan a dan b salah')
    
regula_falsi()
```

### d. Metode Newton

```python
def newton():
    fx = lambda x: x ** 7 - 1000.0
    dfx = lambda x: 7 * (x ** 6)

    x0 = float(input('Masukkan x0: '))
    Er = 100000.0
    max_iterasi = int(input('Masukkan jumlah iterasi maksimum: '))
    k = 0
    
    print('\nk\t\tx(k)\t\tf(x(k))\t\tf\'(x(k))\tx(k+1)\t\tEr')

    while Er > 0.0001 and k < max_iterasi:
        f = fx(x0)
        df = dfx(x0)
        x1 = x0 - (f / df)
        Er = abs((x1 - x0) / x1) * 100
        print('{:d}\t\t{:.6f}\t{:.6f}\t{:.6f}\t{:.6f}\t{:.6f}'.format(k, x0, f, df, x1, Er))
        x0 = x1
        k += 1

    if k == max_iterasi:
        print('Metode Newton tidak konvergen')
    else:
        print('Akar persamaan adalah', round(x1, 6))

newton()
```

### e. Metode Secant

```python
def secant() :
    fx = lambda x: x**7 - 1000
    x0 = float(input('Masukkan x0 :'))
    x1 = float(input('Masukkan x1 :'))
    Er = 10000.0
    max_iterasi = int(input('Masukkan jumlah iterasi maksimum :'))
    k = 0
    
    print('\nk\tx(k-1)\t\tx(k)\t\tf(x(k-1))\tf(x(k))\t\tx(k+1)\t\tEr')
    
    while Er > 0.0001 and k < max_iterasi:
        f0 = fx(x0)
        f1 = fx(x1)
        x2 = x1 - (fx(x1) * (x1 - x0))/(fx(x1) - fx(x0))
        Er = abs((x2 - x1) / x2) * 100
        print('{:d}\t{:.6f}\t{:.6f}\t{:.6f}\t{:.6f}\t{:.6f}\t{:.6f}'.format(k, x0, x1, f0, f1, x2, Er))
        x0 = x1
        x1 = x2
        k += 1
        
    if k == max_iterasi:
        print('\nMetode Newton tidak konvergen')
    else:
        print('\nAkar persamaan adalah', round(x1, 6))
        
secant()
```
